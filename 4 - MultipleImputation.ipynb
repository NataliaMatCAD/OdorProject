```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
odor = pd.read_json('odor_descriptors.json', lines=False, orient='records')
```

```python
odor
```

```python
odor.isna().sum()
```

```python
odor_aux=odor.iloc[:,4:]
odor_aux
```

```python
imp_mean = IterativeImputer(random_state=0)
```

```python
imp_mean.fit(odor_aux)
```

```python
odor_aux=imp_mean.transform(odor_aux)
```

```python
odor_aux
```

```python
len(odor_aux)
```

```python
odor_imput = pd.DataFrame(odor_aux, columns=odor.columns[4:])
```

```python
odor_imput
```

```python
odor_final=pd.concat([odor.iloc[:,:4], odor_imput], axis=1)
```

```python
odor_final.isna().sum()
```
