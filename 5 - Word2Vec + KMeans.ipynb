```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import numpy as np
import matplotlib.pyplot as plt
#from wikipedia2vec import Wikipedia2Vec
```

```python
odor = pd.read_json('Data/odor_relations.json', lines=False, orient='records')
odor_type=odor['name'].value_counts()
odor_type=odor_type[odor_type>38]
print(odor_type.index)
print(len(odor_type.index))
```

```python
odors=odor_type.index
print(len(odors))

wiki2vec=Wikipedia2Vec.load('ModelFile')
memmaps=[]

def get_sentence_vector(word):
	m=0
	for i in word:
		m+=wiki2vec.get_word_vector(str(i))
		return m/len(word)

for i in odors:
	try:
		memmaps.append([i, wiki2vec.get_word_vector(str(i))])
	except:
		memmaps.append([i, get_sentence_vector(i)])

X=[]
for i in range(len(memmaps)):
    X.append(memmaps[i][1])
    
#pd.DataFrame(X).to_csv('X.csv')


tsne=TSNE(n_components=2)
df=tsne.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
label = kmeans.labels_
index = np.unique(label)
for i in index:
    plt.scatter(df[label == i, 0], df[label == i, 1], label=i)
plt.legend()
plt.title('KMeans with 4 clusters')
plt.show()
plt.savefig('KMeans with 4 clusters.png')
label0 = odors[label == 0]
label1 = odors[label == 1]
label2 = odors[label == 2]
label3 = odors[label == 3]
pd.DataFrame(label0).to_csv('label4_0.csv')
pd.DataFrame(label1).to_csv('label4_1.csv')
pd.DataFrame(label2).to_csv('label4_2.csv')
pd.DataFrame(label3).to_csv('label4_3.csv')
```
